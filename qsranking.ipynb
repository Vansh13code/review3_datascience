Q3. Feature Engineering (Using NumPy)


Create a new column Score_Category:

```python
def categorize_score(score):
    if score >= 90:
        return "Elite"
    elif score > 75 and score < 89:
        return "High"
    elif score >= 50 and score <= 74:
        return "Medium"
    else:
        return "Low"
df["Score_Category"]=df["Overall_Score"].apply(categorize_score)
```

## If you are a Policymaker

### 1. What factors should universities improve?
- **Research Quality & Output**
  - Increase high-impact publications
  - Encourage interdisciplinary research
  - Improve citation metrics

- **Faculty Quality**
  - Hire highly qualified and diverse faculty
  - Promote continuous training and development
  - Maintain a strong student–faculty ratio

- **Infrastructure & Technology**
  - Invest in modern labs, libraries, and digital tools
  - Support cloud computing and AI-based learning systems

- **Industry Collaboration**
  - Build partnerships with companies
  - Promote internships, live projects, and placements

- **Global Exposure**
  - Increase international student and faculty intake
  - Encourage exchange programs and collaborations

- **Student Outcomes**
  - Focus on employability and skill-based learning
  - Track alumni success and career growth

---

### 2. Which regions lag behind?
- **Developing Regions (Africa, parts of South Asia)**
  - Limited funding and infrastructure
  - Lower research output

- **Rural/Remote Areas (globally)**
  - Lack of access to quality education and resources
  - Digital divide issues

- **Some Latin American & Southeast Asian regions**
  - Improving but still behind in global rankings due to lower research impact

---

### 3. What trends may shape future rankings?
- **AI & Data-Driven Education**
  - Integration of AI tools in teaching and research

- **Online & Hybrid Learning**
  - Growth of remote education platforms

- **Sustainability & Green Campuses**
  - Focus on environmental impact and SDGs

- **Industry-Relevant Skills**
  - Demand for practical, job-ready skills over theoretical knowledge

- **Global Collaboration**
  - Cross-border research and partnerships

- **Ranking Metrics Evolution**
  - More emphasis on employability, innovation, and societal impact
